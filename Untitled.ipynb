# Forecasting and Modelling

```python
#Focus on only one product for now

from sklearn.ensemble import RandomForestRegressor
import joblib
from datetime import timedelta
import os

serial_nums = df['SerailNum'].unique()
all_data = []

for sn in serial_nums:
    product_df = df[df['SerailNum'] == sn].copy()
    product_df = product_df.set_index('weekend_date').resample('W-SUN').sum().fillna(0)
    product_df['SerailNum'] = sn
    all_data.append(product_df.reset_index())
```

```python
# Fill missing weekly dates

df_full = pd.concat(all_data)
print("Missing weeks filled. New shape:", df_full.shape)
```

```python
# Create features: week, month, and lag values

df_full['week'] = df_full['weekend_date'].dt.isocalendar().week

df_full['month'] = df_full['weekend_date'].dt.month
df_full['lag1'] = df_full.groupby('SerailNum')['quantity'].shift(1)
df_full['lag2'] = df_full.groupby('SerailNum')['quantity'].shift(2)
df_full['rolling_avg3'] = df_full.groupby('SerailNum')['quantity'].shift(1).rolling(3).mean().reset_index(0, drop=True)

```

```python
# Drop rows with NaN after lag creation

df_full.dropna(inplace=True)
```

```python
# train/test split

train_data = df_full[df_full['weekend_date'] < '2024-06-01']
test_data = df_full[(df_full['weekend_date'] >= '2024-06-01') & (df_full['weekend_date'] <= '2024-08-31')]

features = ['week', 'month', 'lag1', 'lag2', 'rolling_avg3']
model_dict = {}
results = []
os.makedirs("models", exist_ok=True)
```

```python
# Train model
for sn in serial_nums:
    train_sn = train_data[train_data['SerailNum'] == sn]
    test_sn = test_data[test_data['SerailNum'] == sn]

    if len(train_sn) < 10 or len(test_sn) == 0:
        continue

    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(train_sn[features], train_sn['quantity'])

    preds = model.predict(test_sn[features])
    actuals = test_sn['quantity'].values
    denom = np.sum(actuals)
    if denom == 0:
        continue  # skip this product because actual sum is zero
    acc = 1 - np.sum(np.abs(preds - actuals)) / denom


    results.append({'SerailNum': sn, 'Validation_Accuracy_Jun_Aug': round(acc, 4)})

    joblib.dump(model, f"models/model_{sn}.pkl")
    model_dict[sn] = model
    
```

```python
# Save validation summary

pd.DataFrame(results).to_csv("validation_accuracy_jun_aug.csv", index=False)
```

```python
monthly_accuracy = []
for month in [6, 7, 8]:
    mdata = test_data[test_data['month'] == month].copy()
    mdata.loc[:, 'predicted'] = mdata.apply(
        lambda row: model_dict[row['SerailNum']].predict(pd.DataFrame([row[features]]))[0] if row['SerailNum'] in model_dict else 0, axis=1)

    actual = mdata.groupby('SerailNum')['quantity'].sum()
    predicted = mdata.groupby('SerailNum')['predicted'].sum()
    both = pd.concat([actual, predicted], axis=1).dropna()
    both.columns = ['actual', 'predicted']

    denom = np.sum(both['actual'])
    if denom == 0:
        continue
    acc = 1 - np.sum(np.abs(both['actual'] - both['predicted'])) / denom
    monthly_accuracy.append({'Month': month, 'Accuracy': round(acc, 4)})

pd.DataFrame(monthly_accuracy).to_csv("monthly_accuracy_jun_jul_aug.csv", index=False)
```

```python
future_forecasts = []
for sn in serial_nums:
    if sn not in model_dict:
        continue

    hist = df_full[df_full['SerailNum'] == sn].copy()
    model = model_dict[sn]

    for _ in range(13):  # forecast next 13 weeks
        last = hist.iloc[-1]
        next_date = last['weekend_date'] + timedelta(weeks=1)

        row = {
            'weekend_date': next_date,
            'week': next_date.isocalendar().week,
            'month': next_date.month,
            'lag1': last['quantity'],
            'lag2': hist.iloc[-2]['quantity'],
            'rolling_avg3': hist.tail(3)['quantity'].mean()
        }

        X_pred = pd.DataFrame([row])[features]
        y_pred = model.predict(X_pred)[0]

        row['quantity'] = y_pred
        row['SerailNum'] = sn
        future_forecasts.append({'SerailNum': sn, 'weekend_date': next_date, 'forecast_quantity': y_pred})

        hist = pd.concat([hist, pd.DataFrame([row])], ignore_index=True)

pd.DataFrame(future_forecasts).to_csv("weekly_forecast_sep_nov.csv", index=False)
```

```python
# Monthly Accuracy Commentary
print("\n--- Monthly Accuracy Commentary ---")
print("We evaluated the model performance for each of the last 3 months (June, July, August 2024)")
print("using the formula: Accuracy = 1 - | - y| / y")
print("Where  is predicted quantity and y is actual quantity.")
print("High accuracy (close to 1) means better forecast.")

print("\nObserved Monthly Accuracies:")
for entry in monthly_accuracy:
    print(f"Month: {entry['Month']}  Accuracy: {entry['Accuracy']}")

print("\nKey Observations:")
print("- Products with consistent weekly patterns performed better.")
print("- Products with zero or highly fluctuating sales had lower accuracy.")
print("- Overall, the model shows reasonable performance with simple ML techniques.")
```
